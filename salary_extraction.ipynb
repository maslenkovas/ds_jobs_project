### Cleaning salary info

```python
with open(path_to_DS_project + 'gcc_jobs_df.pkl', 'rb') as handle:
    jobs_df = pickle.load(handle)
print(jobs_df.shape)
jobs_df.salary_description = jobs_df.salary_description.fillna(-1)
# jobs_with_salary_df = jobs_df[(jobs_df.salary_estimate!=-1)|(jobs_df.salary_description!=-1)]
print('n of rows with salary: ', jobs_df.shape[0])
```

```python
# see if it worked well
i = 0
for job_desription, salary in zip(jobs_df.job_description, jobs_df.salary_description):
    if i > 20: break
    print('----->', salary, '\n', job_desription)
    print('--------------------------------------------------------------------')
    i += 1
```

```python
## clean and organize salary info
# convert types into str
jobs_df.salary_estimate = jobs_df.salary_estimate.astype(str) 
jobs_df.salary_description = jobs_df.salary_description.astype(str) 
# create columns with time period salary info
jobs_df['hourly'] = jobs_df['salary_estimate'].apply(lambda x: 1 if 'per hour' in x.lower() else 0)
# merge two salary columns
jobs_df['salary'] = [a if b=='-1' else b for a,b in zip(jobs_df.salary_estimate.values, jobs_df.salary_description.values)]
# remove unnecessary text info from salary column
jobs_df['salary'] = jobs_df['salary'].apply(lambda x: x.split('(')[0])
# create a column with the currency info 
jobs_df['currency'] = jobs_df['salary'].apply(lambda x: 'usd' if 'usd' in x.lower() or '$' in x.lower() else 'aed' if 'aed' in x.lower()\
                                              else 'bhd' if 'bhd' in x.lower() else 'qar' if 'qar' in x.lower() \
                                              else 'sar' if 'sar' in x.lower() else 'omr' if 'omr' in x.lower() \
                                              else 'kd' if 'kd' in x.lower()else -1)
```

```python
# see the ASCII codes for two types of dashes
print(ord('-'))
print(ord('–'))
```

```python
import re
# get salary column cleaned from text
# remove text from the salary column and replace 'K' with zeroes
salary = jobs_df['salary'].apply(lambda x: x.lower().replace('k', '000'))
salary = salary.apply(lambda x: re.sub('[a-zA-Z$]', '', x))
salary = salary.apply(lambda x: re.sub(' - 0%', '', x))
salary = salary.apply(lambda x: re.sub(',', '', x))
# salary = salary.apply(lambda x: re.sub('.00', '', x))
salary = salary.apply(lambda x: re.sub(' ', '', x))
# create columns for min and max salary
min_salary = salary.apply(lambda x: x.replace('–', '-').split('-')[0])
max_salary = salary.apply(lambda x: x.replace('–', '-').split('-')[1] if len(x.replace('–', '-').split('-'))>1 else x.replace('–', '-').split('-')[0])
jobs_df['min_salary'] = min_salary.astype(float)
jobs_df['max_salary'] = max_salary.astype(float)
# add the average salary column
jobs_df['avg_salary'] = (jobs_df.min_salary + jobs_df.max_salary) / 2
```
